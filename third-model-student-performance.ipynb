```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
data["Extracurricular Activities"]=le.fit_transform(data["Extracurricular Activities"])
```

```python
new_data=data.drop (['Sleep Hours'],axis=1)
```

```python
new_data.head()
```

# split data (train & test)

```python
train_data= new_data.loc[:8000,:]
x_train= train_data.drop (["Performance Index"], axis=1)
y_train= train_data["Performance Index"]
##########
test_data= new_data.loc[8001:,:]
x_test= test_data.drop (["Performance Index"], axis=1)
y_test= test_data["Performance Index"]
```

# Train the mode

```python
from sklearn.linear_model import LinearRegression
le = LinearRegression()
le.fit(x_train,y_train)
```

# acurcy

```python
le.score(x_test,y_test)
```
